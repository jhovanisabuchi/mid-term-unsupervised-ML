```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv(r"C:\Users\johnk\ml-project-unsupervised-learning\Data\clean_data.csv")
df.head()
```

```python
# Step 1: Standardize the data
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
```

```python
# Apply PCA
pca = PCA(n_components=3) 
pca_result = pca.fit_transform(scaled_data)
```

```python
 # Visualize the explained variance ratio
plt.figure(figsize=(8, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_)
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('Explained Variance by Principal Components')
plt.show()
```

```python
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2', 'PC3'])
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming you have the PCA transformed data in a DataFrame `pca_df` with 'PC1', 'PC2', 'PC3' columns
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plotting the first three principal components
ax.scatter(pca_df['PC1'], pca_df['PC2'], pca_df['PC3'], c='blue', marker='o')

# Labels and title
ax.set_title('PCA of Wholesale Data (3 Components)')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')

# Show the plot
plt.show()
```

```python
# Get the loadings (components_)
loadings = pca.components_

loadings_df = pd.DataFrame(loadings, columns=['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen'])

print(loadings_df)
```
